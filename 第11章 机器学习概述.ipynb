#### 1.导入数据

```python
# 从sklearn.datasets库中导入load_iris函数
from sklearn.datasets import load_iris
# 调用load_iris()函数，生成iris数据集对象
iris = load_iris()
# 将前4列的特征数据存入x变量
x = iris.data
# 将最后一列的分类目标数据存入变量y
y = iris.target
```

```python
x[:5]   #查看x的前5行数据
```

```python
y      #查看y的所有数据
```

```python
iris.feature_names    #查看特征数据的名称
```

```python
print(iris.DESCR)    #输出iris数据集的描述信息
```

#### 2.训练模型

```python
#sklearn.neighbors库中导入KNeighborsClassifier函数
from sklearn.neighbors import KNeighborsClassifier
#定义knn模型对象
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x, y)
```

#### 3.模型预测

```python
knn.predict([[2, 5, 4, 2]])
```
